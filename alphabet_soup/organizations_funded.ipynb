## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt


#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df.drop(columns=['EIN', 'INCOME_AMT', 'SPECIAL_CONSIDERATIONS'], inplace=True)
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
cutoff = 500
#create a Boolean mask
application_types_to_replace = application_df['APPLICATION_TYPE'].value_counts()[application_df['APPLICATION_TYPE'].value_counts() < cutoff].index.tolist()

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at NAME value counts for binning
application_df['NAME'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `names_to_replace`
name_cutoff = 100
#create a Boolean mask
names_to_replace = application_df['NAME'].value_counts()[application_df['NAME'].value_counts() < name_cutoff].index.tolist()

# Replace in dataframe
for name in names_to_replace:
    application_df['NAME'] = application_df['NAME'].replace(name,"Other")

# Check to make sure binning was successful
application_df['NAME'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
application_df["CLASSIFICATION"].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
# create a Boolean mask
application_df["CLASSIFICATION"].value_counts()[application_df["CLASSIFICATION"].value_counts() > 1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
class_cutoff = 1000
#create a Boolean mask
classifications_to_replace = application_df['CLASSIFICATION'].value_counts()[application_df['CLASSIFICATION'].value_counts() < class_cutoff].index.tolist()

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
categorical_cols = application_df.dtypes[application_df.dtypes == 'object'].index.tolist()

dummies_application_df = pd.concat([pd.get_dummies(application_df[categorical_cols]),
    application_df.drop(categorical_cols, axis=1)], axis=1)

dummies_application_df
```

```python
# Split our preprocessed data into our features and target arrays
y = dummies_application_df.IS_SUCCESSFUL
X = dummies_application_df.drop(columns='IS_SUCCESSFUL')

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

nn = tf.keras.models.Sequential()

# ReLU - Rectified Linear Unit 
nn.add(tf.keras.layers.Dense(units=64, activation='relu', input_dim=63))

# First hidden layer
nn.add(tf.keras.layers.Dense(units=32, activation='swish'))

#additional layer
nn.add(tf.keras.layers.Dense(units=24, activation='tanh'))


# Second hidden layer
nn.add(tf.keras.layers.Dense(units=16, activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=["accuracy"])
```

```python
# Train the model
nn_fit = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
